```python
import json
from tqdm import tqdm
```

```python
input_path = "../data/final_dataset.jsonl"

data = []
with open(input_path, "r") as f:
    for line in f:
        data.append(json.loads(line))

print("Total samples:", len(data))
```

```python
processed_data = []

for item in tqdm(data):
    messages = item["messages"]
    
    user_msg = ""
    assistant_msg = ""
    
    for m in messages:
        if m["role"] == "user":
            user_msg = m["content"]
        elif m["role"] == "assistant":
            assistant_msg = m["content"]
    
    if len(user_msg) > 5 and len(assistant_msg) > 5:
        processed_data.append({
            "instruction": user_msg,
            "input": "",
            "output": assistant_msg
        })

print("Converted samples:", len(processed_data))
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"\s+", " ", text)
    return text.strip()

for item in processed_data:
    item["instruction"] = clean_text(item["instruction"])
    item["output"] = clean_text(item["output"])

print("Cleaning done")
```

```python
unique_data = []
seen = set()

for item in processed_data:
    key = item["instruction"] + item["output"]
    if key not in seen:
        seen.add(key)
        unique_data.append(item)

print("After dedup:", len(unique_data))
```

```python
import random
random.shuffle(unique_data)
print("Shuffled")
```

```python
output_path = "../data/final_cleaned.jsonl"

with open(output_path, "w") as f:
    for item in unique_data:
        f.write(json.dumps(item) + "\n")

print("Final dataset ready!")
```
